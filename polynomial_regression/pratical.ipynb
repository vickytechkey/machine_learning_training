```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('poly_data.csv')
```

```python
df.info()
```

```python
sns.scatterplot(data=df, x='X', y='y',color='red')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df[['X']],df['y'],test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
```

```python
x_poly_train = poly.fit_transform(x_train)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_poly_train,y_train)
```

```python
sns.scatterplot( x=x_poly_train[:,1], y=model.predict(x_poly_train), color='blue')
```

```python
x_poly_test = poly.transform(x_test)
sns.scatterplot( x=x_poly_test[:,1], y=model.predict(x_poly_test), color='green')
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, model.predict(x_poly_test))
r2 = r2_score(y_test, model.predict(x_poly_test))   
```

```python
print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
```
