```python
def pass_fail(x):
    if x == 1:
        return 'Pass'
    else:
        return 'Fail'
```

```python
import joblib
model = joblib.load('logistic_regression_model_balanced.pkl')
```

```python
input_data = [[55, 78, 45, 56, 67]]
prediction = model.predict(input_data)
result = pass_fail(prediction[0])
```

```python
print("The student is predicted to:", result)
```
