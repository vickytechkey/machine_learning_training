```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Sample data
x1 = np.array([1, 2, 3])
x2 = np.array([1, 0, 1])
y = np.array([20, 30, 50])

# Predicted model: y = 10 + 15*x1 + 0*x2
def predict(x1, x2):
    return 10 + 15*x1 + 0*x2

# Create grid
x1_grid, x2_grid = np.meshgrid(np.linspace(0, 4, 10), np.linspace(0, 2, 10))
y_pred = predict(x1_grid, x2_grid)

# 3D plot
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

# Plot training points
ax.scatter(x1, x2, y, color='r', label='Training points')

# Plot plane
ax.plot_surface(x1_grid, x2_grid, y_pred, alpha=0.5)

ax.set_xlabel('x1 (hours studied)')
ax.set_ylabel('x2 (sleep hours)')
ax.set_zlabel('y (marks scored)')
ax.legend()
plt.show()
```
